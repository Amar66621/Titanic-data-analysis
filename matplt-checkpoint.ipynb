```python
import matplotlib.pyplot as plt
import numpy as np
```

Lineplot

```python
days=list(range(1,16))
```

```python
days
```

```python
temp=[20.0,21,19.5,23.0,24,25.2,12,28,32,56,45,20,15,10.0,25]
```

```python
len(temp)
```

```python
plt.figure(figsize=(5,3))
plt.plot(days,temp)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(days,temp)
plt.title('weather report', color='blue', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(days,temp)
plt.title('weather report', color='blue', fontsize=15)
plt.xlabel('No. of days', color='red', fontsize=13)
plt.ylabel('Temperature', color='red', fontsize=13)
plt.show()
```

# hist plot

```python
marks=np.random.randint(25,90,100)
```

```python
marks
```

```python
len(marks)
```

```python
plt.figure(figsize=(5,3))
plt.hist(marks)
plt.show()
```

```python
bins=[40,50,60,70,80]
```

```python
plt.figure(figsize=(5,3))
plt.hist(marks,bins=bins,color='r',rwidth=0.8,label='marks of maths')
plt.show()
```

```python
x=['maths','physics', 'chemistry', 'english']
y=[35,56,48,90]
```

# pie

```python
plt.pie(y,labels=x)
plt.show()
```

```python
color=['r','m','b','g']
explode=[0,0.1,0,0]
plt.pie(y,labels=x,colors=color,explode=explode, startangle=45)
plt.show()
```
